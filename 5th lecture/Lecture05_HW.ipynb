**2. Построить бинарный классификатор для каждого семейства(Family) на основе LogisticRegression c $l_1$ и $l_2$ регуляризацими. Это соответствует целевым исходам (Конкретное семейство, Всё остальное). Подобрать гиперпараметр $C$ с помощью кросс-валидации, используя метрику ROC AUC.**

```python
from sklearn.linear_model import LogisticRegressionCV
from sklearn.model_selection import cross_validate, cross_val_score


res = []
df_train = df.drop(['Family','Genus','Species'], axis = 1)
df_trainmm = df_train.apply(lambda x: (x-min(x))/(max(x)-min(x)))
for value in df['Family'].unique():
    target = df['Family'] == value
    cv_res = []
    for C in range(100,200,2):
        clf = LogisticRegressionCV(Cs = C, penalty = 'l1')
        cv = cross_val_score(clf, df_trainmm, target, n_jobs=-1, cv=3, scoring='roc_auc')
        cv_res.append((C, cv.mean()))
        print(cv.mean())
    df_cv = pd.DataFrame(cv_res, columns = ['C', 'cv.mean'])
    maxcv = df_cv.sort_values(by = ['cv.mean'], ascending = False).head(1).values
    #print(maxcv)
    res.append((value, maxcv[0], maxcv[1]))
df_fam = pd.DataFrame(res, columns=['Family', 'C', 'max cv'])
##df['Family'].unique
```
