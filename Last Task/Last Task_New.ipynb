```python
#Dropping ID and SalePrice
test_id = test_set['Id']
target = train_set['SalePrice']
test_set.drop(['Id'], axis = 1, inplace=True)
train_set.drop(['Id', 'SalePrice'], axis = 1, inplace=True)
```

Let's divide our columns into two groups

```python
num_cols = []
cat_cols = []

for col in train_set.columns:
    if train_set[col].dtype == 'float64' or train_set[col].dtype == 'int64':
        num_cols.append(col)
    elif train_set[col].dtype == 'object':
        cat_cols.append(col)
```

Let's work with data for a bit

```python
# Заполним пропуски в числовых полях - средним
def medianReplacer(df):
    res = df.fillna(df.mean())
    return res

for col in num_cols:
    medianReplacer(train_set[col])
    medianReplacer(test_set[col])
train_set.isnull().sum().sort_values()
```

And with cat cols

```python
def catColsNan(df):
    res = df.fillna(df.value_counts().idxmax())
    return res


for col in cat_cols:
    catColsNan(train_set[col])
    catColsNan(test_set[col])
```

```python
train_set.isnull().sum().sort_values()
```

```python
for col in ('GarageType', 'GarageFinish', 'GarageQual'):
    train_set[col] = train_set[col].fillna('NoGRG')
    test_set[col] = test_set[col].fillna('NoGRG')
for col in ('BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2'):
    train_set[col] = train_set[col].fillna('NoBSMT')    
    test_set[col] = test_set[col].fillna('NoBSMT') 
train_set['FireplaceQu'] = train_set['FireplaceQu'].fillna('NoFP')
test_set['FireplaceQu'] = test_set['FireplaceQu'].fillna('NoFP')

train_set['Alley'] = train_set['Alley'].fillna('NoAlley')
test_set['Alley'] = test_set['Alley'].fillna('NoAlley')
```

```python
def custmDrop(df):
    res = df
    for col in df.columns:
        if (len(df[col].unique()) == 1 or df[col].isna().mean()>0.5):
            res = res.drop(col,inplace=False,axis=1)
    return res


train_set = custmDrop(train_set)
test_set = custmDrop(test_set)
```

Сделаем one-hot кодирование

```python
new_train = pd.DataFrame()
new_test = pd.DataFrame()
for col in num_cols:
    new_train[col] = train_set[col]
    new_test[col] = test_set[col]

# OneHot

for col in cat_cols:
    oh1 = pd.get_dummies(train_set[col], prefix = col, dtype=int)
    oh2 = pd.get_dummies(test_set[col], prefix = col, dtype=int)
    new_train = pd.concat([new_train, oh1], axis = 1)
    new_test = pd.concat([new_test, oh2], axis = 1)

new_train.head()    
```
