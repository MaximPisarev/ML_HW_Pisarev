```python
import zipfile
import numpy as np
import pandas as pd
#Считаем файл трейна и теста в таблицу пандас
z_train = zipfile.ZipFile("../data/training.zip")
train = pd.read_csv(z_train.open("training.csv"))

print(train.shape)
train.head()
```

```python
target = train['Label']
train_new = train.drop(columns = ['Weight', 'Label', 'EventId'])
train_new.head()
```

Начнем с Наивного Байесса 

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
#from sklearn.model_selection import train_test_split
#
#train1, test1, target_train, target_test = train_test_split(train_new, target, shuffle=False)

gnb = GaussianNB()
bnb = BernoulliNB()


gnb.fit(train_new, target)
bnb.fit(train_new, target)

acc_gnb = gnb.score(train_new, target)
acc_bnb = bnb.score(train_new, target)

print(acc_gnb)
print(acc_bnb)

```

Теперь деревья

```python
%%time
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

minitrain = train_new[:10000]
minitarget = target[:10000]

params = {
    'max_depth' : np.arange(2, 20, 1),
    'criterion' : ['gini', 'entropy'],
    'max_features' : np.arange(1, train_new.shape[1], 1),
    'min_samples_leaf' : np.arange(2, 30, 1)
}


tree = DecisionTreeClassifier()

grid = GridSearchCV(
    tree, params, scoring='roc_auc',n_jobs=-1,
    cv=5, refit = True
)
grid.fit(minitrain, minitarget)
print(grid.best_params_)
print(grid.best_score_)
```

KNN 

```python
%%time
from sklearn.neighbors import KNeighborsClassifier

minitrain = train_new[:10000]
minitarget = target[:10000]
params = {
    'n_neighbors' : np.arange(2, 20, 1),
    'algorithm' : ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'leaf_size' : np.arange(25, 55, 5),
    'p' : [1, 2]
}

knn = KNeighborsClassifier()


grid_knn = GridSearchCV(
    knn, params, scoring='roc_auc',n_jobs=-1,
    cv=5, refit = True
)

grid_knn.fit(minitrain, minitarget)
print(grid_knn.best_params_)
print(grid_knn.best_score_)
```

Линейная Регрессия 

```python
%%time
from sklearn.linear_model import LogisticRegression


params = {
    'penalty' : ['l1', "l2", "elasticnet", "none"],
    'solver' : ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
    'max_iter' : np.arange(500, 10000, 500)
}



lr = LogisticRegression()

grid_lr= GridSearchCV(
    lr, params, scoring='roc_auc',n_jobs=-1,
    cv=5, refit = True
)

grid_lr.fit(minitrain, minitarget)
print(grid_lr.best_params_)
print(grid_lr.best_score_)

```

Рандомные Деревья

```python
%%time
from sklearn.ensemble import RandomForestClassifier

params = {
    'n_estimators' : np.arange(50, 500, 50),
    'criterion' : ['gini', 'entropy'],
    'max_depth' : np.arange(2, 20, 1),
    'min_samples_leaf' : np.arange(2, 20, 1),
}


Rf = RandomForestClassifier()


grid_rf = GridSearchCV(
    Rf, params, scoring='roc_auc',n_jobs=-1,
    cv=5, refit = True
)

grid_rf.fit(minitrain, minitarget)
print(grid_rf.best_params_)
print(grid_rf.best_score_)
```

Stacking

```python
from sklearn.ensemble import StackingClassifier
from sklearn.model_selection import cross_validate
#MiniStacking
clfs = [GaussianNB, DecisionTreeClassifier(criterion = 'gini', max_depth = 7, max_features = 25, min_samples_leaf = 23)
       ,KNeighborsClassifier(algorithm = 'auto', leaf_size = 25, n_neighbors = 19, p = 1),
       LogisticRegression(max_iter = 9500, penalty = 'l1', solver = 'liblinear'),
       RandomForestClassifier(criterion = 'entropy', max_depth = 17, min_samples_leaf = 3, n_estimators = 450)]


stck = StackingClassifier(estimators = clfs, n_jobs = -1, cv = 5)

stck.fit(minitrain,minitarget)

acc = stck.score(minitrain,minitarget)
```
